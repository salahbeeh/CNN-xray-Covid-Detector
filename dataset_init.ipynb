```python
import pandas as ps
import os
import shutil
```

```python
# collecting data for positve patients form source1 (github)

dataPath= "covid-chestxray-dataset-master/metadata.csv"
rawImgsPath= "covid-chestxray-dataset-master/images"
```

```python
# discoving the dataset

df = ps.read_csv(dataPath)
print (df.shape)
```

so we have 950 rows and 30 columns 

```python
df.head(5)
```

Data shows that we have data of positve and negative corona patiants, we need to extract the positive ones

```python
covidDir = "covidData/covid"

if not os.path.exists(covidDir):
    os.mkdir(covidDir)
    print("covid folder created")
```

```python
counter = 0

for (i,row) in df.iterrows():
    if row["finding"]== "Pneumonia/Viral/COVID-19":
        counter+=1
        
print(counter)        
        
```

so we have 584 positve case in the dataset provided form the github repo

```python
df["view"]
```

going throgh the following link, I found that there is many views for a single X-Ray the pa is the frontend of it and what we baisclly looking for
https://en.wikipedia.org/wiki/Chest_radiograph

```python
counter = 0

for (i,row) in df.iterrows():
    if row["finding"]== "Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        imageName = row["filename"]
        imagePath = os.path.join(rawImgsPath,imageName)
        imgCopyPath = os.path.join(covidDir,imageName)
        shutil.copy2(imagePath,imgCopyPath)
        print("Transfar image",counter)
        counter+=1
        
print(counter) 
```

so we got about 196 pic of a frontend chest x-rays

```python
# collecting some of the non-covid images form the source 2 (kaggle)

import random

kagglePath = "chest_xray kaggle/train/NORMAL"
nonCovidDir = "covidData/nonCovid"

imgNames = os.listdir(kagglePath)
random.shuffle (imgNames)
```

```python
nonCovidDir = "covidData/nonCovid"

if not os.path.exists(nonCovidDir):
    os.mkdir(nonCovidDir)
    print("covid folder created")



for i in range (196):
    imgName = imgNames[i]
    imgPath = os.path.join(kagglePath,imgName)
    imgCopyPath = os.path.join(nonCovidDir,imgName)
    shutil.copy2(imgPath,imgCopyPath)
    print("Transfar image",i)
```
